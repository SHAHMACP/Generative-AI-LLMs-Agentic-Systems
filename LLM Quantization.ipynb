```python
from transformers import AutoModelForCausalLM
import torch
model_name = "microsoft/phi-2"
def get_model_size(model):
    param_size = 0
    for param in model.parameters():
        param_size += param.nelement() * param.element_size()
    for name, param in model.named_parameters():
      print(f"dtype: {param.dtype}")
      break
    return param_size / (1024**2)  # Convert bytes to MB
```

1. Load model in full precision (FP32)

```python
print("Loading FP32 model...")
model_fp32 = AutoModelForCausalLM.from_pretrained(model_name,token="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")
size_fp32 = get_model_size(model_fp32)
print(f"FP32 model size: {size_fp32:.2f} MB")
```

2. Load 8-bit quantized model

```python
print("\nLoading 8-bit quantized model...")
model_8bit = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_8bit=True,
    device_map='auto',
    token="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"
)
size_8bit = get_model_size(model_8bit)
print(f"8-bit quantized model size: {size_8bit:.2f} MB")
```

Load 4-bit quantized model

```python
print("\nLoading 4-bit quantized model...")
model_4bit = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_4bit=True,
    device_map='auto',
    token="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"
)
size_4bit = get_model_size(model_4bit)
print(f"4-bit quantized model size: {size_4bit:.2f} MB")
```
